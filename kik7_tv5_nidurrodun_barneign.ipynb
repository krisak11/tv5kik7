**Verkefnalýsing**

Hjúkrunarfræðideild þarf að raða niður öllum nemendum sem eru á fjórða ári í hjúkrun í bókleg námskeið og í klíník. Klíník er starfsnám þar sem nemendur þurfa að vinna á stofnun eða spítala undir handleiðslu kennara á svæðinu.

Tveir áfangar eru með það fáa í hóp í einu að klíníkin þar þarf að dreifast yfir heilt ár og því þarf að fylgjast vel með að allir komist í klíník og klári allt sem þarf til að útskrifast. Þessir áfangar eru Barnahjúkrun og Bráðahjúkrun. 

Áfangar sem er skylt er að taka á  fjórða ári eru:

**Haust:**

*   Heimahjúkrun - þarf klíník sem klárast haustið 
*   Hjúkrunarstjórnun - þarf klíník sem klárast um haustið
*   Langveikir og heilbrigðisþjónustan - eingöngu bóklegur áfangi haustið
*   Ofbeldi og geðrækt - eingöngu bóklegur áfangi sem klárast um haustið
*   Bráðahjúkrun - bóklegur og klínískur áfangi þar sem bóklegi hlutinn klárast um haustið en klíníkin byrjar um haustið og lýkur um vorið
*   Barnahjúkrun - bóklegur og klínískur áfangi þar sem bóklegi hlutinn klárast um haustið en klíníkin byrjar um haustið og lýkur um vorið

Klínískt nám má byrja í 35 viku (29.08.2022 og standa yfir í 6 vikur) þá tekur við bóklegt nám til viku 45 (7.11.2022 og er í 3 vikur). Á þeim þrem vikum má koma fyrir tveim hópum í Barnahjúkrun, tveim í Bráðahjúkrun og tveim í Heimahjúkrun. 

**Vor:**

Nemendur fá að velja sér tvo valáfanga af fimm sem eru í boði. Þeir eru:

*   Kynheilbrigði 
*   Konur, heilsa og samfélag

*   Andvari: Gjörgæsluhjúkrun

```python
#fasti sem skilgreinir staðsetningu í datetime_lista þar sem bráðahjúkrun_val byrjar)
eftir_brada_val = 27
# útfærsla á líkani með Gurobi
namskeid = klinik_fjoldi.index.tolist()

m = Model("Klínik") # frumstilla Gurobi líkan

x = m.addVars([(s,c,d,k) for s in nemendur for c in namskeid for d in dagsetning for k in deild[c]], 
              vtype=GRB.BINARY)

## HÉR KEMUR ÚTFÆRSLAN YKKAR ##

#setjum markfall fyrir soft skorðu, til að hámarka nemendur úti á landi.
m.setObjective(quicksum(x[s,'Heimahjúkrun HJÚ711G',d,'landsbyggd'] for s in utalandi for d in dagsetning), GRB.MAXIMIZE)
#setjum markfall fyrir soft skorðu, til að lágmarka að þessir nemendur sem eru í barneignaleyfi séu skráðir í 
#námskeið á þeim tíma.
m.setObjective(quicksum(x[29,c,d,k] for c in namskeid  for k in deild[c] for d in forbidden_29), GRB.MINIMIZE)
m.setObjective(quicksum(x[57,c,d,k] for c in namskeid for k in deild[c] for d in forbidden_57),GRB.MINIMIZE)
m.setObjective(quicksum(x[68,c,d,k] for c in namskeid for k in deild[c] for d in forbidden_68),GRB.MINIMIZE)


# Hér er dæmi um skorðu: athuga hvort nemandi sé skráður í þetta námskeið (af 4 klinik namskeid).
m.addConstrs(
    quicksum(x[s,c,d,k] for d in dagsetning for k in deild[c]) 
    == nemendur_skraning.loc[s,c] for s in nemendur for c in namskeid
    )

#  ef nemandi er skráður í valnámskeið á þessum degi þá má hann ekki vera í klíník á sama degi
m.addConstrs(quicksum(x[s,c,d,k]  for n in valnamskeid_nemendur[s]
                      for d in stundata[n] for c in namskeid for k in deild[c]) == 0 for s in nemendur)

#fjöldatakmörk í klínik : fjöldi nemenda í ákveðinni deild <= hámarks fjöldi nemenda í klíník á þessari deild á ákveðinni dagsetningu
m.addConstrs(quicksum(x[s,c,d,k] for k in deild[c] for s in nemendur) <= klinik_fjoldi.loc[c,d] for c in namskeid for d in dagsetning)

#nemandi er bara í hámarki einum klíník_áfanga á hverri dagsetningu [lota] 
m.addConstrs(quicksum(x[s,c,d,k] for c in namskeid for k in deild[c]) <= 1 for s in nemendur for d in dagsetning)

#nemandi má ekki vera í bráðahjúkrun_val í sömu viku og bráðahjúkrun_klínik eða eftir.
forbidden_dagsetning = ['3/14/2022', '3/21/2022']
m.addConstrs(quicksum(x[s,'Bráðahjúkrun HJÚ824G',d,k] for k in deild['Bráðahjúkrun HJÚ824G'] for d in forbidden_dagsetning) == 0 
             for s in nemendur if 'Bráðahjúkrun HJÚ0AVM' in valnamskeid_nemendur[s])

#Vorum að hugsa að gera þetta svona en náðum ekki að fá það til að virka:
#----- dynamic útfærsla af skorðunni fyrir ofan. notar date í stað string, og ætti að útiloka allar dagsetningar
## sem koma eftir bráðahjúkrun valnámskeiðið. -- virkar ekki ennþá ----------
#m.addConstrs(quicksum(x[s,'Bráðahjúkrun HJÚ824G',d,k] for k in deild['Bráðahjúkrun HJÚ824G'] for d in date_list if d >= date_list[eftir_brada_val] ) == 0 
#             for s in nemendur if 'Bráðahjúkrun HJÚ0AVM' in valnamskeid_nemendur[s])

#ef nemandi s er skráður í 'Hjúkrunarstjórnun HJÚ714G' þá þarf hann að þekkja deildina (sést í stjornun_deild_nemendur)
m.addConstrs(quicksum(x[s,'Hjúkrunarstjórnun HJÚ714G',d,k] for d in dagsetning for k in stjornun_deild_nemendur[s]) 
             == 1 for s in nemendur if nemendur_skraning.loc[s, 'Hjúkrunarstjórnun HJÚ714G'] == 1)
#setjum harða skorðu að nemandi 29, 57, og 68 sem er í bareignaleyfi á ákveðnum tíma má ekki vera skráður í 
#námskeið á þeim dögum.
m.addConstrs(quicksum(x[29,c,d,k] for c in namskeid for k in deild[c]) == 0
             for d in forbidden_29)
#m.addConstrs(quicksum(x[57,c,d,k] for c in namskeid for k in deild[c]) == 0
#             for d in forbidden_57)
m.addConstrs(quicksum(x[68,c,d,k] for c in namskeid for k in deild[c]) == 0
             for d in forbidden_68)
m.optimize()
```
